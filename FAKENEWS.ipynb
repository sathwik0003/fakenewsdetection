```python
#Lemmatizition and stemming(partially)

import re


def lemmatize(word):
    # some of my own lemmatization rules
    lemmatization_rules = {
        "running": "run",
        "ran": "run",
        "flies": "fly",
        "children": "child",
        "apply": "apply",
        "slider": "slide",
        'mice': 'mouse',
        'better': 'good',
        'fattest': 'fat',
        'larger': 'large',
        'largest': 'large',
        'fatter': 'fat',
        'flew': 'fly',
        'frozen': 'freeze',
        'threw': 'throw',
        'caught': 'catch',
        'swam': 'swim',
        'ate': 'eat',
        'went': 'go',
        'wrote': 'write',
        'danced': 'dance',
        'bought': 'buy',
        'men': 'man',
        'women': 'woman',
        'feet': 'foot',
        'teeth': 'tooth',
        'lice': 'louse',
        'bacteria': 'bacterium',
        'octopi': 'octopus',
        'criteria': 'criterion',
        'analyses': 'analysis',
        'syllabi': 'syllabus',
        'phenomena': 'phenomenon',
        'theses': 'thesis',
        'crises': 'crisis',
        'foci': 'focus',
        'oases': 'oasis',
        'lives':'life',
        'knives':'knife',
        'dying':'die',
        'dead': 'die',
        'cried':'cry',
        'lied':'lie',
        'laid':'lay',
        'lying':'lie',
        'took':'take',
        'brought': 'bring',
        'fallen':'fall',
        'straighten':'straight',
        'liking':'like',
        'hiking': 'hike',
        'smoking':'smoke',
        'faking':'fake',
        'joking':'joke',
        'making': 'make',
        'poking':'poke',
        'dating':'date',
        'rose':'rise',
        'gone':'go',
        'taught':'teach',
        'emerging': 'emerge',
        'bouncing':'bounce',
        'glancing':'glace',
        'colliding':'collide',
        'agreed':'agree',
        'arraging':'arrange',
        'defining':'define',
        'analysing': 'analyse',
        'restoring': 'restore',
        'excluding':'exclude',
        'gloves':'glove',
    }

    # Check if the word is in the  lemmatization rules
    if word in lemmatization_rules:
        return lemmatization_rules[word]
    
    #threaten--> threat
    #lighten-->light
    #brighten-->bright
    
    if word.endswith('ten'):
         return re.sub(r'en$','',word)
        
        
    #driven-->drive
    #given-->give
    #taken-->take
    
    if word.endswith('en'):
        return re.sub(r'en$','e',word)
    
    
    #Sleepier--> sleepy
    #uglier--> ugly
    #naughtier--> naughty
    #nastier--> nasty
    
    if word.endswith('ier'):
        return re.sub(r'ier$', 'y', word)
    
    
    #leaves--> leaf
    #believes--> belief
    #claves--> calf
    #elves--> elf
    #halves-->half
    
    if word.endswith('ves'):
        return re.sub(r'ves$', 'f', word)

    ## problem with words ending with e
    # liking will be like but drinking is simply drink
    suffixes = [ 'es', 'ed', 'ing', 'er','s']

    for suffix in suffixes:
        if word.endswith(suffix):
            return re.sub(f'{suffix}$', '', word)
        
        #if nothing matches return the word itself
    return word
check_word = input("enter a word")
display(lemmatize(check_word))
```

```python
import re

def extract_context(sentence, target_word):
    tokens = re.findall(r'\w+|\$[\d\.]+', sentence)
    index = tokens.index(target_word)
    context_words = tokens[max(0, index - 2):index] + tokens[index + 1:min(len(tokens), index + 6)]
    return context_words

def word_sense_disambiguation(word, context):
    senses = {
         'election': {
        'senses': {
            'political event': ['vote', 'candidate', 'campaign'],
            'electrical process': ['circuit', 'voltage', 'current']
        },
        'default': 'political event'
    },
    'corruption': {
        'senses': {
            'political corruption': ['bribery', 'embezzlement', 'scandal'],
            'physical deterioration': ['decay', 'rot', 'deterioration']
        },
        'default': 'political corruption'
    },
    'protest': {
        'senses': {
            'political demonstration': ['demonstration', 'rally', 'activism'],
            'express objection': ['oppose', 'object', 'disapprove']
        },
        'default': 'politcal demonstration'
    },
    'policy': {
        'senses': {
            'government strategy': ['strategy', 'plan', 'decision'],
            'insurance agreement': ['coverage', 'premium', 'beneficiary']
        },
        'default': 'government strategy'
    },
    'scandal': {
        'senses': {
            'political scandal': ['corruption', 'misconduct', 'investigation'],
            'public outrage': ['outrage', 'controversy', 'disgrace']
        },
        'default': 'political scandal'
    },
    'legislation': {
        'senses': {
            'political laws': ['law', 'act', 'regulation'],
            'act of making laws': ['process', 'procedure', 'legal system']
        },
        'default': 'political laws'
    },
    'diplomacy': {
        'senses': {
            'international relations': ['foreign policy', 'negotiation', 'diplomat'],
            'tact and skill in dealing with people': ['tact', 'skill', 'negotiation']
        },
        'default': 'international relations'
    },
    'partisanship': {
        'senses': {
            'bias towards a political party': ['bias', 'prejudice', 'loyalty'],
            'partiality or attachment to a cause': ['support', 'devotion', 'dedication']
        },
        'default': 'bias towards a political party'
    },
    'debate': {
        'senses': {
            'formal discussion': ['discussion', 'argument', 'deliberation'],
            'contentious argument': ['dispute', 'controversy', 'clash']
        },
        'default': 'formal discussion'
    },
    'campaign': {
        'senses': {
            'political campaign': ['election', 'candidate', 'strategy'],
            'organized course of action': ['effort', 'drive', 'initiative']
        },
        'default': 'political campaign'
    },
    'proclamation': {
        'senses': {
            'official announcement': ['announcement', 'declaration', 'statement'],
            'act of declaring something': ['declaration', 'assertion', 'announcement']
        },
        'default': 'offical announcement'
    },
    'policyholder': {
        'senses': {
            'insurance holder': ['insured', 'beneficiary', 'coverage'],
            'government policy gainer': ['benfit', 'supporter', 'adherent','benfitiary']
        },
        'default': 'government policy gainer'
    },
    'censorship': {
        'senses': {
            'control of information': ['restriction', 'suppression', 'control'],
            'exclusion of inappropriate content': ['ban', 'forbid', 'restriction']
        },
        'default': 'exclusive of inappropriate content'
    },
    'constituency': {
        'senses': {
            'voters in a specified area': ['voters', 'electorate', 'district'],
            'group of supporters or customers': ['supporters', 'customers', 'base']
        },
        'default': 'voters in a specified area'
    },
    'lobbying': {
        'senses': {
            'effort to influence decisions': ['influence', 'advocacy', 'pressure'],
            'seeking to influence others': ['persuasion', 'campaigning', 'negotiation']
        },
        'default': 'effort to influence others'
    },
    'endorsement': {
        'senses': {
            'public support or approval': ['support', 'approval', 'recommendation'],
            'formal declaration of support': ['declaration', 'support', 'backing']
        },
        'default': 'formal declaration of support'
    },
    'policy-making': {
        'senses': {
            'process of creating policies': ['decision-making', 'formulation', 'implementation'],
            'creating a set of principles': ['principles', 'guidelines', 'doctrine']
        },
        'default': 'creating set of principles'
    },
    'propaganda': {
        'senses': {
            'biased information to promote a cause': ['promotion', 'manipulation', 'bias'],
             'dissemination of ideas': ['information', 'communication', 'spread']
        },
        'default': 'biased information to promte a cause'
    },
}

    # Check if the word is in the senses dictionary
    if word in senses:
        # Iterate over senses and check context overlap
        for sense, context_words_for_sense in senses[word]['senses'].items():
            overlap = set(context) & set(context_words_for_sense)
            if overlap:
                return sense  # Return the first sense with context overlap

        # Check if 'default' key exists before accessing it
        if 'default' in senses[word]:
            return senses[word]['default']

    # Return None if the word is not in the senses dictionary
    return None

# Example sentence
sentence = "The is a political propaganda."
word_to_disambiguate = "propaganda"

# Extract context words
context_words = extract_context(sentence, word_to_disambiguate)

print(f"Context words: {context_words}")

# Perform WSD
sense = word_sense_disambiguation(word_to_disambiguate, context_words)

# Print the correct default sense
print(f"The sense of '{word_to_disambiguate}' in the given context is: {sense or (senses[word_to_disambiguate]['default'] if 'default' in senses[word_to_disambiguate] else 'default')}")
```

```python
import random

def train_test_split(features, labels, test_size=0.2, random_state=None):
    if random_state is not None:
        random.seed(random_state)

    data = list(zip(features, labels))
    random.shuffle(data)

    split_index = int(len(data) * (1 - test_size))

    train_data = data[:split_index]
    test_data = data[split_index:]

    train_features, train_labels = zip(*train_data)
    test_features, test_labels = zip(*test_data)

    return train_features, test_features, train_labels, test_labels



train_data, test_data, train_labels, test_labels = train_test_split(
    news['text'],  
    news['result'], 
    test_size=0.2,  
    random_state=42
)

train_samples = list(zip(train_data, train_labels))


def get_bi_grams(tokens):
    bi_grams = [(tokens[i], tokens[i + 1]) for i in range(len(tokens) - 1)]
    return bi_grams

# Create a dictionary to store bi-gram frequencies for fake and genuine news
fake_bi_grams = {}
genuine_bi_grams = {}


for tokens, label in train_samples:
    bi_grams = get_bi_grams(tokens)
    if label == 0: 
        for bi_gram in bi_grams:
            fake_bi_grams[bi_gram] = fake_bi_grams.get(bi_gram, 0) + 1
    else:  
        for bi_gram in bi_grams:
            genuine_bi_grams[bi_gram] = genuine_bi_grams.get(bi_gram, 0) + 1



def predict_label(document):
    bi_grams = get_bi_grams(document)
    fake_score = sum(fake_bi_grams.get(bi_gram, 0) for bi_gram in bi_grams)
    genuine_score = sum(genuine_bi_grams.get(bi_gram, 0) for bi_gram in bi_grams)

    # Predict the label based on the higher score
    return 0 if fake_score > genuine_score else 1

def calculate_accuracy(true_labels, predicted_labels):
    correct_predictions = sum(1 for true, predicted in zip(true_labels, predicted_labels) if true == predicted)
    total_samples = len(true_labels)
    accuracy = correct_predictions / total_samples
    return accuracy

# Make predictions on the test data
predictions = [predict_label(document) for document in test_data]

# Calculate accuracy
accuracy = calculate_accuracy(test_labels, predictions)
print(f"Accuracy: {accuracy:.2%}")

```
